### Counting number of black and white and coloured picturs in the data set

```python
import cv2
from os import listdir
from os.path import isfile, join


onlyfiles = ['../../data/images/' + f for f in listdir('../../data/images') if isfile(join('../../data/images', f))]

d = {}
for f in onlyfiles:
    im = cv2.imread(f, cv2.IMREAD_UNCHANGED)
    if str(im.shape) not in d.keys():
        d[str(im.shape)] = 1
    else:
        d[str(im.shape)] += 1
d
```

### Occurrencies of classes

### Counting number of classes per each image

```python
import cv2
from os import listdir
from os.path import isfile, join
import re


onlyfiles = [int(re.sub("[^0-9]", "", f)) for f in listdir('../../data/images') if isfile(join('../../data/images', f))]
onlyfiles.sort()
d = {}
annotations = ['../../data/annotations/' + f for f in listdir('../../data/annotations') if isfile(join('../../data/annotations', f))]
for item in onlyfiles:
    d[item] = 0
    for annotation in annotations:
        with open(annotation, 'r') as f:
            for line in f.readlines():
                if str(item) == line.strip():
                    d[item] += 1

d
```

### Grouping images by number of classes in it

```python
import pandas as pd

df = pd.DataFrame.from_dict(d, orient='index')
df['classes'] = df[0]
df.groupby('classes').count()
```
